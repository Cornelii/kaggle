# Single feature distribution

```python
def single_distrib(feature_name,ax,df,hue=None):
    ## countplot
    sns.countplot(x=feature_name,data=df,ax=ax,hue=hue)
    ax.set_title(feature_name)
    
def plot_single_distrib(df,feature_list,row=1,col=1,figsize=(13,8),hue=None):
    fig,ax = plt.subplots(row,col,figsize=figsize)
    if row == 1 and col == 1:
        for feature in feature_list:
            single_distrib(feature,ax,df,hue)
            
    elif row == 1 and col != 1:
        for jmi,feature in enumerate(feature_list):
            single_distrib(feature,ax[jmi],df,hue)
    else:
        for jmi,feature in enumerate(feature_list):
            widx,hidx=divmod(jmi,col)
            single_distrib(feature,ax[widx][hidx],df,hue)
```

```python
feature_list=Categorical_features+Number_features
plot_single_distrib(df_train,feature_list,4,2,(15,30))
```

```python
feature_list=Categorical_features+Number_features
plot_single_distrib(df_train,feature_list,4,2,(15,30),"Survived")
```
