```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=4)

%matplotlib inline
```

```python
df_train=pd.read_csv('./data/train.csv')
df_test=pd.read_csv('./data/test.csv')
```

# Null data check

## trains set

```python
for col in df_train.columns:
    percent=100*(df_train[col].isnull().sum()/df_train[col].shape[0])
    msg="column: {}\t Percent of NaN value: {:.2f}".format(col,percent)
    print(msg)
```

## test set

```python
for col in df_test.columns:
    percent=100*(df_test[col].isnull().sum()/df_test[col].shape[0])
    msg="column: {}\t Percent of NaN value: {:.2f}".format(col,percent)
    print(msg)
```

```python
fig,axs=plt.subplots(1,2,figsize=(18,8))

df_train["Survived"].value_counts().plot.pie(ax=axs[0])
sns.countplot("Survived",data=df_train,ax=axs[1])
```
